```python
## Header (importing dependencies)
from __future__ import print_function
import pandas as pd

import math

from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials
from gspread_dataframe import set_with_dataframe, get_as_dataframe

## Authentication
gc = gspread.authorize(GoogleCredentials.get_application_default())

## Importing data frame
base_spreadsheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/18MsDQvexK68Thn7Sx9SEa8GI8b-I9W3575iYVc5j3CQ/edit?usp=sharing')
worksheet = base_spreadsheet.sheet1
dataFrame = pd.DataFrame(worksheet.get_all_records())

## Variables
n_students = 24
n_classes = 60
```

```python
## Averaging
mean = (dataFrame['P1'] + dataFrame['P2'] + dataFrame['P3'])/3
for i in range(n_students):
  if mean[i] >= 70:
    dataFrame.loc[i,'Situação'] = 'Aprovado'
  elif mean[i] < 70 and mean[i] >= 50:
    dataFrame.loc[i,'Situação'] = 'Exame Final'
  else:
    dataFrame.loc[i,'Situação'] = 'Reprovado por nota'

dataFrame
```

```python
## Checking the absences
for i in range(n_students):
  if dataFrame.loc[i, 'Faltas'] >= (n_classes*0.25):
    dataFrame.loc[i,'Situação'] = 'Reprovado por falta'
dataFrame
```

```python
## Computing naf
for i in range(n_students):
  if dataFrame.loc[i, 'Situação'] == 'Exame Final':
    dataFrame.loc[i,'Nota para Aprovação Final'] = math.ceil(100 - mean[i])
  else:
    dataFrame.loc[i,'Nota para Aprovação Final'] = 0
dataFrame
```

```python
## Overwriting the original data frame 
set_with_dataframe(worksheet, dataFrame)
dataFrame
```
